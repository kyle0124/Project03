```python
def first_name(name) :
    for i in range(len(name.split(' '))) :
        if 'ix' in name.split(' ')[i] :
            tmp = ' '.join(name.split(' ')[:i+1])
        elif '투싼' in name.split(' ')[i] :
            tmp = ' '.join(name.split(' ')[:i+1])
    return tmp
```

```python
df['first_name'] = df['name'].apply(lambda x : first_name(x))
df
```
