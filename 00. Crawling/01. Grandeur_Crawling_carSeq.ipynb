```python
src_list = []
url1 = 'https://www.kbchachacha.com/public'
url2 = '/search/main.kbc#!?sort=-orderDate&makerCode=101&classCode=1101&'
url3 = 'carCode=3301,3302,3048,3049,2852,2879,1411,1486,1106,1412,1105&gas=004005,004003,004002,004001&page='


for i in range(20, 40) :
    url = url1 + url2 + url3
    url = url + str(i + 1)
    driver.get(url)
    html = driver.page_source
    soup = bs4.BeautifulSoup(html, 'html.parser')
    src1 = soup.find('div', class_='cs-list02 cs-list02--ratio small-tp generalRegist')
    src2 = list(src1.find_all('button', class_="lc-btn like ck-item __wish"))
        
    for src in src2 :
        src_list.append(str(src).split('\"')[3])
        
src_list
```
